```python

# 做完特徵工程前的所有準備
import pandas as pd
import numpy as np
import datetime
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score

from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#匯入資料
data_path = 'data-science-london-scikit-learn/'
df_train = pd.read_csv(data_path + 'train.csv', header=None)
df_test = pd.read_csv(data_path + 'test.csv', header=None)
df_trainLabels = pd.read_csv(data_path + 'trainLabels.csv', header=None,names=['class'])
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
df_trainLabels
```

```python
#二元分類
print( df_trainLabels['class'].nunique() )
print(np.sum(df_trainLabels['class'] ==1 ))
print(np.sum(df_trainLabels['class'] ==0 ))
```

```python
# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(df_train, df_trainLabels, test_size=0.25, random_state=4)
```

```python
# 建立模型
clf = DecisionTreeClassifier()

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)


acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
#print("Feature importance: ", clf.feature_importances_)
```

```python
#GradientBoosting
# 建立模型
GBclf = GradientBoostingClassifier(n_estimators=100)

# 訓練模型
GBclf.fit(x_train, y_train)

# 預測測試集
y_pred = GBclf.predict(x_test)


acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ", acc)
```

```python
submit = GBclf.predict(df_test)
```

```python
index = [i for i in range(1, len(submit)+1)]
submit = pd.DataFrame({'Solution': submit})
submit['Id'] = np.arange(1,submit.shape[0]+1)
submit = submit[['Id','Solution']]
```

```python
submit.to_csv('london.csv', index = False)
```
